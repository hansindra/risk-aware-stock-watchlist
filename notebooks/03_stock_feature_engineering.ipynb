## Stock Feature Engineering

This section transforms the cleaned stock dataset into a feature-rich dataset that can be used for anomaly detection.

The goal of feature engineering is to create meaningful numerical variables that describe stock behaviour and market context. These features are designed to capture:

- daily movement
- rolling risk
- short-term trend
- downside risk
- overall market behaviour
- stock movement relative to the market

The final output of this section is a clean feature dataset saved as `stock_features_daily.csv`, which is later used for anomaly detection and evaluation.

```python
# Load Clean Stock Dataset 

from pathlib import Path
import pandas as pd
import numpy as np

def find_project_root(start: Path) -> Path:
    start = start.resolve()
    for p in [start] + list(start.parents):
        if (p / "data").exists():
            return p
    return start

PROJECT_ROOT = find_project_root(Path.cwd())
DATA_DIR = PROJECT_ROOT / "data"
DATA_PROCESSED = DATA_DIR / "processed"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATA_PROCESSED exists:", DATA_PROCESSED.exists())

df = pd.read_csv(DATA_PROCESSED / "stocks_master_2014_2024.csv")

df["Date"] = pd.to_datetime(df["Date"])

df = df.sort_values(["Ticker","Date"]).reset_index(drop=True)

print("Shape:", df.shape)
df.head()
```

```python
#  Daily Stock Return

df["Return"] = (
    df.groupby("Ticker")["Close"]
      .pct_change()
)

df.head(10)
```

```python
#30-Day Stock Volatility

df["Vol_30"] = (
    df.groupby("Ticker")["Return"]
      .rolling(30)
      .std()
      .reset_index(level=0, drop=True)
)

df.head(40)
```

```python
#20-Day Momentum

df["Mom_20"] = (
    df.groupby("Ticker")["Close"]
      .pct_change(20)
)

df.head(30)
```

```python
#Drawdown

rolling_max = (
    df.groupby("Ticker")["Close"]
      .cummax()
)

df["Drawdown"] = (df["Close"] / rolling_max) - 1

df.head(30)
```

```python
# Load ASPI Data (Correct Path)

ASPI_PATH = PROJECT_ROOT / "data" / "raw" / "aspi" / "ASPI_2014_2024_raw.csv"

aspi = pd.read_csv(ASPI_PATH)

aspi["Date"] = pd.to_datetime(aspi["Date"])

aspi = aspi.sort_values("Date").reset_index(drop=True)

print("ASPI shape:", aspi.shape)
print("ASPI min date:", aspi["Date"].min())
print("ASPI max date:", aspi["Date"].max())

aspi.head()
```

```python
#Proxy Market Return (Equal Weighted)

market_proxy = (
    df.groupby("Date")["Return"]
      .mean()
      .reset_index()
      .rename(columns={"Return": "Market_Return"})
)

print("Proxy market shape:", market_proxy.shape)
market_proxy.head()
```

```python
# Proxy Market 30-Day Volatility

market_proxy["Market_Vol_30"] = (
    market_proxy["Market_Return"]
      .rolling(30)
      .std()
)

market_proxy.head(40)
```

```python
#  Merge Proxy Market to Stock Data

df = df.merge(market_proxy, on="Date", how="left")

print("Shape after merge:", df.shape)
df.head()
```

```python
# Relative Return


df["Rel_Return"] = df["Return"] - df["Market_Return"]

df.head(10)
```

```python
# Drop Initial NaNs


feature_cols = [
    "Return",
    "Vol_30",
    "Mom_20",
    "Drawdown",
    "Market_Return",
    "Market_Vol_30",
    "Rel_Return"
]

df_clean = df.dropna(subset=feature_cols).copy()

print("Original rows:", len(df))
print("After dropna:", len(df_clean))

df_clean.head()
```

```python
# Save Clean Feature Dataset


OUTPUT_PATH = DATA_PROCESSED / "stock_features_daily.csv"

df_clean.to_csv(OUTPUT_PATH, index=False)

print("Saved successfully.")
print("Final shape:", df_clean.shape)
```
