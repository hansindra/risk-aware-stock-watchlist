### Check Processed Data Files

This code locates the `data/processed` folder and lists all available files.  
It is used to verify that preprocessing outputs have been successfully generated.

```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path.cwd().parents[0]
DATA_PROCESSED = BASE_DIR / "data" / "processed"

print("Files in processed:")
for f in DATA_PROCESSED.iterdir():
    print(" -", f.name)
```

### Inspect Stock Dataset

This code loads the master stock dataset and checks:
- available columns  
- number of unique stocks  
- date range  
- sector distribution  

This can verify that the dataset is complete and correctly structured.

```python
stocks = pd.read_csv(DATA_PROCESSED / "stocks_master_2014_2024.csv")

print(stocks.columns)
print("Tickers:", stocks["Ticker"].nunique())
print("Date range:", stocks["Date"].min(), stocks["Date"].max())
print(stocks.groupby("Sector")["Ticker"].nunique())
```
